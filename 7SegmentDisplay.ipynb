```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#Load Data

```python
df = pd.read_csv('data/7SegmentDisplay.csv')
```

```python
df
```

#Transform Data

```python
X = df.loc[:, df.columns != 'Digit']
X
```

```python
Y = pd.get_dummies(df['Digit'], prefix='Digit')
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
x_train = np.array(X)
```

```python
y_train = np.array(Y)
```

```python
x_train
```

```python
y_train
```

#Create Model

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, Adam
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(7,), activation='relu'))
model.add(Dense(10, activation='softmax'))
model.compile(Adam(lr=0.05), 'categorical_crossentropy', metrics=['accuracy'])
```

#Fit Model

```python
model.fit(x_train, y_train, epochs=500, verbose=1)
```

#Model Prediction Test

```python
x_test = np.array([[ 1.,  1.,  1.,  1.,  1.,  1.,  0.]])
```

```python
y = model.predict(x_test)
```

```python
y
```

#Save Model

```python
model.summary()
```

```python
model.save('7SegmentDisplay_model.h5')
```

#Export CoreML Model

```python
from keras.models import load_model  
model = load_model('7SegmentDisplay_model.h5')
```

```python
import coremltools
```

```python
coreml_model = coremltools.converters.keras.convert(model, input_names="leds", output_names="digits")
```

```python
coreml_model.save('7SegmentDisplay_model.mlmodel')
```
